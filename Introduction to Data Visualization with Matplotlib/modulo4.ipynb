```python
import matplotlib.pyplot as plt
import pandas as pd

df =  pd.read_csv('../datasets/seattle_weather.csv')
seattle_weather = df[:12] # so queremos trabalhar com dado de um ano
seattle_weather['MONTH'] = seattle_weather['DATE'].apply(lambda x: pd.Timestamp(month=x, day=1, year=2000).strftime('%b'))

# fazendo analogo 
df =  pd.read_csv('../datasets/austin_weather.csv')
austin_weather = df[:12] # so queremos trabalhar com dado de um ano
austin_weather['MONTH'] = austin_weather['DATE'].apply(lambda x: pd.Timestamp(month=x, day=1, year=2000).strftime('%b'))
```

## Estilo de grafico 

Vamos usar **plt.style.use**  repare varios grafico abaixo para ver a diferen√ßa

Link para saber mais estilo [https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

```python
plt.style.use("bmh")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

```python
plt.style.use("ggplot")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

```python
plt.style.use("default")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

# Automa√ß√£o 

Aqui esta como sento, tipo nos n√£o precisamos colocar o nome do esporte e criar para cada coluna, estamos automatizando nesse sentindo 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summer_2016_medals = pd.read_csv('../datasets/summer2016.csv')
```

```python
sports = summer_2016_medals["Sport"].unique()
print(sports)
```

```python
fig, ax = plt.subplots()
for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
    ax.bar(sport, sport_df["Height"].mean(),
    yerr=sport_df["Height"].std())
    

ax.set_ylabel("Height (cm)")
ax.set_xticklabels(sports, rotation=90)
plt.show()
```
